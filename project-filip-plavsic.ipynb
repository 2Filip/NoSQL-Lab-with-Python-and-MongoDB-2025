## Summary

This project explores the Game of Thrones quotes dataset and enriches it with metadata from the “An API of Ice and Fire” API. The workflow follows a complete ELT (Extract–Load–Transform) pipeline — from connecting to MongoDB, cleaning and loading data, to transforming and analyzing quotes by character and house. The aim is to uncover patterns in language use, quote length, and thematic content across the fictional universe.

```python
# download all characters in one call
r = requests.get(API_URL, timeout=20); r.raise_for_status()
data = r.json()

# inspect first character
character = data[0].copy()
# hide None values (same trick as template)
for k in list(character.keys()):
    if character[k] is None:
        del character[k]
character
```

```python
# normalize helpers
def norm_character(c):
    house = c.get("house") or {}
    _id   = c.get("slug") or sha(c.get("name",""))
    return {
        "_id": _id,
        "name": c.get("name"),
        "slug": c.get("slug"),
        "house_id": house.get("slug"), # explicit id for joins
        "house_name": house.get("name"),
        "house_slug": house.get("slug"),
        "aliases": c.get("aliases", []),
        "titles":  c.get("titles", []),
    }

def norm_quotes_from_character(c):
    ch = norm_character(c)
    out = []
    for text in c.get("quotes", []):
        out.append({
            "_id": sha(text + ch["_id"]),
            "text": text,
            "speaker_id": ch["_id"],
            "speaker_name": ch["name"],
            "house_id": ch.get("house_id"),
            "house_name": ch.get("house_name"),
            "house_slug": ch.get("house_slug"),
        })
    return ch, out
```

```python
# loop all characters → two lists
list_chars, list_quotes = [], []
for c in data:
    ch, qs = norm_quotes_from_character(c)
    list_chars.append(ch)
    list_quotes.extend(qs)

len(list_chars), len(list_quotes)
```

```python
# derive distinct houses from characters
houses_raw = {}
for ch in list_chars:
    if ch.get("house_id"):
        hid = ch["house_id"]
        houses_raw[hid] = {
            "_id": hid,
            "slug": hid,
            "name": ch.get("house_name"),
            "members": set()
        }
        houses_raw[hid]["members"].add(ch["_id"])

# finalize docs
list_houses = []
for h in houses_raw.values():
    h["members"] = sorted(list(h["members"]))
    h["member_count"] = len(h["members"])
    list_houses.append(h)

len(list_houses), list_houses[:2]
```

### Additional implementation of An API of Ice And Fire

```python
import re
import requests

AOIF = "https://anapioficeandfire.com/api/houses"
session = requests.Session()

def _normalize(s: str) -> str:
    if not s:
        return ""
    # normalize curly quotes etc.
    s = s.replace("’", "'").replace("‘", "'").replace("“", '"').replace("”", '"')
    s = re.sub(r"\s+", " ", s).strip()
    return s

def _to_query_variants(name: str):
    "Generate a few plausible queries for AoIF."
    name = _normalize(name)
    if not name:
        return []

    # exact form, e.g. "House Stark of Winterfell"
    variants = [name]

    # remove 'House ' prefix (AoIF sometimes stores both ways)
    if name.lower().startswith("house "):
        variants.append(name[6:].strip())

    # remove ' of ...' suffix → "House Stark"
    if " of " in name.lower():
        variants.append(name.split(" of ")[0].strip())

    # last token (e.g., 'Stark' from 'House Stark of Winterfell')
    last = name.split()[-1]
    variants.append(last)

    # dedupe
    seen, uniq = set(), []
    for v in variants:
        v2 = v.lower()
        if v2 not in seen:
            uniq.append(v)
            seen.add(v2)
    return uniq

def aoif_fetch_by_name(name: str):
    "Try several variants, then paginate/fuzzy match as a fallback."
    if not name:
        return None

    # Try several exact-ish requests
    for q in _to_query_variants(name):
        try:
            r = session.get(AOIF, params={"name": q}, timeout=25)
            if r.ok and r.json():
                return r.json()[0]
        except Exception:
            pass

    # Fallback: paginate and leniently look for any variant token in AoIF name
    variants = [v.lower() for v in _to_query_variants(name)]
    page, found = 1, None
    while page <= 5 and not found:
        try:
            r = session.get(AOIF, params={"page": page, "pageSize": 50}, timeout=25)
            if not r.ok:
                break
            for item in r.json():
                nm = _normalize(item.get("name", "")).lower()
                # If any variant token appears in the AoIF name, accept it
                if any(v in nm for v in variants):
                    found = item
                    break
            page += 1
        except Exception:
            break
    return found
```

```python
import time

for h in list_houses:
    meta = aoif_fetch_by_name(h.get("name"))
    if not meta:
        continue
    h["region"]     = meta.get("region")
    h["words"]      = meta.get("words")
    h["coatOfArms"] = meta.get("coatOfArms")
    h["seats"]      = [s for s in meta.get("seats", []) if s]
    time.sleep(0.25)
```

### Load

```python
# insert lists into MongoDB
from pymongo import UpdateOne

# de-dup in memory by _id
def dedup_by_id(docs):
    tmp = {}
    for d in docs:
        tmp[d["_id"]] = d
    return list(tmp.values())

list_chars  = dedup_by_id(list_chars)
list_quotes = dedup_by_id(list_quotes)

# UPSERTS (idempotent)
if list_chars:
    res_ch = chars.bulk_write(
        [UpdateOne({"_id": d["_id"]}, {"$set": d}, upsert=True) for d in list_chars],
        ordered=False
    )
    print(f"characters upserted: {res_ch.upserted_count}, matched: {res_ch.matched_count}")

if list_quotes:
    res_qt = quotes.bulk_write(
        [UpdateOne({"_id": d["_id"]}, {"$set": d}, upsert=True) for d in list_quotes],
        ordered=False
    )
    print(f"quotes upserted: {res_qt.upserted_count}, matched: {res_qt.matched_count}")

print("counts:", chars.count_documents({}), quotes.count_documents({}))
```

#### Dynamic Hierarchical Structure: Region --> House --> Character

We visualize the nested relationships from regions to houses to characters, showing the full structure of the GoT data. This not only serves as a data model demonstration but also enables analytical insight into affiliations and structure.

```python
import matplotlib.pyplot as plt

def plot_tree(nodes, parent, x=0, y=0, dx=1.0, level=1, pos={}, connections=[]):
    pos[parent] = (x, y)
    children = [n["_id"] for n in nodes if n.get("parent") == parent]
    width = dx / max(1, len(children))
    for i, child in enumerate(children):
        cx = x + (i - (len(children) - 1) / 2.0) * width
        cy = y - 1
        connections.append((parent, child))
        plot_tree(nodes, child, cx, cy, width, level+1, pos, connections)
    return pos, connections

nodes = list(db.house_lineage.find())
roots = [n["_id"] for n in nodes if n.get("parent") is None]
plt.figure(figsize=(10, 6))
for i, root in enumerate(roots):
    pos, connections = plot_tree(nodes, root, x=i*3, y=0, dx=2)
    for (p, c) in connections:
        if p is not None:
            px, py = pos[p]
            cx, cy = pos[c]
            plt.plot([px, cx], [py, cy], 'k-')
    for node, (x, y) in pos.items():
        plt.scatter(x, y, s=2000, color='#C3E6FC')
        plt.text(x, y, node, ha='center', va='center', fontsize=10, weight='bold')
plt.title("Lineage Tree of GoT Houses and Members")
plt.axis('off')
plt.show()
```

### Datastructure

#### UML class diagram

![Class diagram](../images/Class_diagram.png)

This class diagram visually describes the structure of the MongoDB database for Game of Thrones quotes analytics. The model includes:
- House: Represents a noble house, with attributes like name, region, words (motto), and references to its members.
- Character: Each character is uniquely identified and linked to a House via house_id. Characters can have aliases and titles.
- Quote: Each quote record contains the text, length, speaker information, and is linked to both the speaker (a Character) and their House.
- Quote_Enriched: An enhanced version of a quote, containing additional NLP-derived attributes such as tokens, type-token ratio, themes, and a question flag.
- House_Metrics: Aggregated metrics for each house, such as number of quotes, average quote length, and list of speakers, referenced by the house’s id.

Relationships:
- One House has many Characters (1-to-0..*)
- One Character speaks many Quotes (1-to-0..*), and each Quote is spoken by exactly one Character (shown by speaker_id)
- One House contains many Quotes (indirectly, via Character and Quote)
- Each Quote may have one Enriched Quote (1-to-0..1)
- Each House may have one House_Metrics record (1-to-0..1)

Foreign key fields are shown as associations, and multiplicities reflect the true cardinalities from your data model.

### Characters with the richest vocabulary (type/token ratio)

```python
res = list(quotes.aggregate([
  {"$match": {"words": {"$exists": True}}},
  {"$group": {"_id": "$speaker_name", "bags": {"$push": "$words"}}},
  {"$project": {
     "uniq": {"$size": {"$setUnion": "$bags"}},
     "total": {"$sum": {"$map": {"input": "$bags", "as": "b", "in": {"$size":"$$b"}}}}
  }},
  {"$addFields": {"ttr": {"$cond": [{"$gt": ["$total", 0]}, {"$divide": ["$uniq", "$total"]}, 0]}}},
  {"$sort": {"ttr": -1}}, {"$limit": 10}
]))
pd.DataFrame(res)
```

### Theme share by character

```python
res = list(db.quotes.aggregate([
  {"$set":
   {"theme": 
    {"$regexMatch": {"input": "$text", 
                     "regex":"(king|queen|dragon|winter)",
                     "options":"i"}}}},
    {"$group": {"_id":"$speaker_name", 
                "q_total":{"$sum":1}, 
                "q_theme":{"$sum": {"$cond":["$theme",1,0]}}}},
    {"$set":{"theme_share":
             {"$divide":["$q_theme","$q_total"]}}},
  {"$sort":{"theme_share":-1}}, {"$limit":15}
]))
pd.DataFrame(res)
```

### Mentions of themes

### Average words per quote by character

```python
res = list(db.quotes.aggregate([
  {"$match": {"words": {"$exists": True}}},
  {"$addFields": {"word_count": {"$size": "$words"}}},
  {"$group": {"_id": "$speaker_name",
              "avg_words": {"$avg": "$word_count"},
              "quotes": {"$sum": 1}}},
  {"$match": {"quotes": {"$gte": 3}}},  # avoid 1-quote outliers
  {"$sort": {"avg_words": -1}},
  {"$limit": 15}
]))
pd.DataFrame(res)
```

### Co-occurrence of words inside quotes

```python
res = list(db.quotes.aggregate([
  {"$match": {"words": {"$exists": True}}},
  {"$project": {"pairs": {
      # make small windows of word pairs (skip identical/doubles)
      "$reduce": {
        "input": {"$range": [0, {"$size": "$words"}]},
        "initialValue": [],
        "in": {"$concatArrays": ["$$value", {
          "$map": {
            "input": {"$slice": ["$words", {"$add": ["$$this", 1]}, 4]},
            "as": "w2",
            "in": {"$concat": [
              {"$arrayElemAt": ["$words", "$$this"]}, "|", "$$w2"
            ]}
          }
        }]}
      }
  }}},
  {"$unwind": "$pairs"},
  {"$group": {"_id": "$pairs", "w": {"$sum": 1}}},
  {"$sort": {"w": -1}},
  {"$limit": 40}
]))
pd.DataFrame(res[:10])
```
